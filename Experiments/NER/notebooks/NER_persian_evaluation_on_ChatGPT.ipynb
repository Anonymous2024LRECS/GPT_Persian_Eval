```python
def get_completion(prompt, model="gpt-3.5-turbo"):   #  TODO:  use gpt-4   https://platform.openai.com/docs/guides/chat
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
        # max_tokens=256
    )
    return response.choices[0].message["content"]
```

```python

l1 = "[('طاهایی', 'PER'), ('گفت', 'O'), (':', 'O'), ('0', 'O'), ('طرح', 'O'), ('عمرانی', 'O'), ('،', 'O'), ('0', 'O'), ('طرح', 'O'), ('مدرسه\u200cسازی', 'O'), ('در', 'O'), ('قالب', 'O'), ('0', 'O'), ('کلاس', 'O'), ('درس', 'O'), ('،', 'O'), ('پنج', 'O'), ('طرح', 'O'), ('مسکن', 'Product'), ('و', 'O'), ('چند', 'O'), ('طرح', 'O'), ('فرهنگی', 'Event'), ('دیگر', 'O'), ('از', 'O'), ('جمله', 'O'), ('طرحهائی', 'O'), ('است', 'O'), ('كه', 'O'), ('در', 'O'), ('هفته', 'O'), ('دولت', 'ORG'), ('مورد', 'O'), ('بهره\u200cبرداری', 'O'), ('قرار', 'O'), ('می\u200cگیرند', 'O'), ('.', 'O')]"
def convert_2_list(l1):
  l1 = l1.replace('\'',"\"")
  # print(l1)
  res = l1.strip('][').split(', ')
  out = []
  for i in range(0,len(res)-1,2):
    a = res[i].replace('\"','')
    a = a.replace('(','')
    a = a.replace(')','')
    b = res[i+1].replace('\"','')
    b = b.replace('(','')
    b = b.replace(')','')
    out.append((a,b))
  return out
convert_2_list(l1)
```

```python
def comparable_gold_pred(response, ls, ts):
  res = convert_2_list(response)
  # print(response)
  # print(res)
  # print(ts)
  # print([i_2_tags[i] for i in ls])
  i2l = [i_2_tags[i] for i in ls]
  for i, v in enumerate(ts):
    # print(v)
    # if i < len(res):
      # print(res[i])
    # else:
    #   res.append(("", ''))
    if i >= len(res):
      res.append(("", ''))
      # print(res[i])
    # print(i2l[i])
    # print(res[i][0].replace("\\u200c"," "), ts[i].replace("\u200c"," "))
    if res[i][0].replace("\\u200c"," ") == ts[i].replace("\u200c"," "):
      if res[i][1].lower() in i2l[i].lower():
        # print(i, 'match', i2l[i], 'word', ts[i])
        tempi = list(res[i])
        tempi[1] = i2l[i]
        res[i] = tuple(tempi)
    else:
      # print('yyyyyy')
      # print(res, i2l)
      tempi = list(res[i])
      tempi[1] = 'O'
      res[i] = tuple(tempi)
      continue
  # print('res len', len(res), 'inp len', len(ts), 'labels len', len(ls))
  # print([i[1] for i in res])
  # print(i2l)
  return [i[1] for i in res], i2l
```

```python
def report(golds_bn, y_pred_bn):
  pdic = {i:0 for i in i_2_tags.keys()}
  gdict = {i:0 for i in i_2_tags.keys()}
  cpdict = {i:0 for i in i_2_tags.keys()}

  for i in golds_bn:
    for j in i:
      gdict[j] += 1

  for i in y_pred_bn:
    for j in i:
      pdic[j] += 1

  for i,v in enumerate(y_pred_bn):
    for j,l in enumerate(v):
      if j < len(golds_bn[i]):
        if l == golds_bn[i][j]:
          cpdict[golds_bn[i][j]] += 1


  res = 0
  for i in i_2_tags.keys():
    if gdict[i] != 0: # i != 0 and
      prec = 0 if pdic[i] == 0 else (cpdict[i]/pdic[i])
      rec = 0 if gdict[i] == 0 else (cpdict[i]/gdict[i])
      res += 0 if rec+prec == 0 else (2*prec*rec)/(rec+prec)

  return res/len(i_2_tags.keys())
```
