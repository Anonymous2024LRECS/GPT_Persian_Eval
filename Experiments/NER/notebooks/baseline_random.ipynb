```python
from datasets import load_dataset

dataset = load_dataset("persian_ner", split="test") # "Khedesh/ArmanNER"   "test" "train"
```

```python
tags = ["O", "I-event", "I-fac", "I-loc", "I-org", "I-pers", "I-pro", "B-event", "B-fac", "B-loc", "B-org", "B-pers", "B-pro"]
tags_2_i = {v:i for i,v in enumerate(tags)}
i_2_tags = {i:v for i,v in enumerate(tags)}
```

```python
import random

preds = []
golds = []
ic = 0

for i in dataset:
  ts = i['tokens']
  ls = i['ner_tags']

  print(ts)
  print(ls)

  if ts == []:
    continue


  pred = [random.choice(list(range(len(tags)))) for i in ts]
  gold = ls

  print(pred)
  print(gold)
  print(gold[0])

  preds.append(pred)
  golds.append(gold)

  ic += 1
  if ic > 201:
    break
```

```python
def report(golds_bn, y_pred_bn):
  pdic = {i:0 for i in i_2_tags.keys()}
  gdict = {i:0 for i in i_2_tags.keys()}
  cpdict = {i:0 for i in i_2_tags.keys()}

  for i in golds_bn:
    for j in i:
      gdict[j] += 1

  for i in y_pred_bn:
    for j in i:
      pdic[j] += 1

  for i,v in enumerate(golds_bn):
    for j,l in enumerate(v):
      if l == y_pred_bn[i][j]:
        cpdict[y_pred_bn[i][j]] += 1

  print(cpdict)
  print(pdic)
  print(gdict)
  res = 0
  for i in i_2_tags.keys():
    if gdict[i] != 0:  # i != 0 and
      prec = 0 if pdic[i] == 0 else (cpdict[i]/pdic[i])
      rec = 0 if gdict[i] == 0 else (cpdict[i]/gdict[i])
      res += 0 if rec+prec == 0 else (2*prec*rec)/(rec+prec)

  return res/len(i_2_tags.keys())
```

```python
report(golds, preds)
```
